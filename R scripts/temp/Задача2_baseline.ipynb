```python
import pandas as pd
import numpy as np
```

```python
clf_data = pd.read_csv('C:/Trees/xtrain.csv', sep=',')
clf_target = pd.read_csv('C:/Trees/ytrain.csv', sep=',')
```

```python
for i in clf_data:
    print(i, clf_data[i].nunique())
```

```python
cat_var = clf_data.loc[:, clf_data.apply(pd.Series.nunique) < 30].columns
for i in cat_var:
    clf_data[i] = clf_data[i].astype('object')
    
clf_data.info()
```

```python
for i in cat_var:
     print(clf_data[i].value_counts(dropna=False))
```

```python
for i in cat_var:
    clf_data[i] = np.where(clf_data[i].isnull(), '9999', clf_data[i])
```

```python
for i in cat_var:
     print(clf_data[i].value_counts(dropna=False))
```

```python
for i in cat_var:
    clf_data.loc[clf_data[i].value_counts()[clf_data[i]].values < 1000, i] = 'Rare_value'
```

```python
for i in cat_var:
     print(clf_data[i].value_counts(dropna=False))
```

```python
num_columns = [c for c in clf_data.columns if clf_data[c].dtype.name != 'object']
for i in num_columns:
    clf_data[i].fillna(9999, inplace=True)
```

```python
# Разделим выборку на обучение и тест.
from sklearn.model_selection import train_test_split
clf_train_data, clf_test_data, clf_train_labels, clf_test_labels = train_test_split(clf_data, clf_target,
                                                                                    test_size=0.3, random_state=1)
```

```python
clf_train_data = pd.get_dummies(clf_train_data)
clf_test_data = pd.get_dummies(clf_test_data)
```

```python
print(clf_train_data.shape)
print(clf_train_labels.shape)
print(clf_test_data.shape)
print(clf_test_labels.shape)
```

```python
clf_train_data = clf_train_data.values
clf_test_data = clf_test_data.values
clf_train_labels = clf_train_labels.values 
clf_test_labels = clf_test_labels.values
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
```

```python
xgb = XGBClassifier(n_estimators=50, learning_rate=0.02, 
                    max_depth=20, subsample=0.9, 
                    colsample_bytree=1, colsample_bylevel=0.9, 
                    reg_alpha=0.1, reg_lambda=0.1, random_state=0, n_jobs=8, verbose=1)
xgb.fit(clf_train_data,clf_train_labels)
# печатаем значения правильности
print("AUC на обучающей выборке: {:.3f}".
      format(roc_auc_score(clf_train_labels, xgb.predict_proba(clf_train_data)[:, 1])))
print("AUC на контрольной выборке: {:.3f}".
      format(roc_auc_score(clf_test_labels, xgb.predict_proba(clf_test_data)[:, 1])))
```
